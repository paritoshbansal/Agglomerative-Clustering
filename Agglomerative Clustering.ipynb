```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/DataScience75/batch89/main/german_credit_data.csv"
data=pd.read_csv(url)
```

```python
data.head()
```

```python
cluster_data=data[['Age','Credit amount', 'Duration']]
```

```python
cluster_data.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agm=AgglomerativeClustering(n_clusters=2,linkage="ward")   ### linkage = "complete", "ward" ,"single"
```

```python
agm.fit(cluster_data)
```

```python
pred=agm.fit_predict(cluster_data)
```

```python
pred
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(11,6))
dnd = sch.dendrogram(sch.linkage(cluster_data,method="ward"))
```

```python
cluster_data["Cluster"]=pred
cluster_data.head()
```
